#### 10. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
#In this case I would erase the values because they all return zero. 
#But if the data were relevant and did not exceed 70% of null data, I would surely keep them.
low_variance = []


for col in merge._get_numeric_data():
    minimum = min(merge[col])
    ninety_perc = np.percentile(merge[col], 90)
    if ninety_perc == minimum:
        low_variance.append(col)
        
print(low_variance)
```
